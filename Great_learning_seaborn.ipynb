```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
# we ploting the line plot based on 'event'
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event') # grouping by 'hue'
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='region') # grouping by 'hue'
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='subject') # grouping by 'hue'
plt.show()
```

```python
# adding style
sns.lineplot(x='timepoint',y='signal',data=fmri,
             hue='event',
             style='event')
plt.show()
```

```python
# here we mad line style
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='subject',style='subject')
plt.show()
```

```python
# here we find up's and down's 
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event', markers=True)
plt.show()
```
