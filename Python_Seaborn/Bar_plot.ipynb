```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
pokemon=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/pokemon.csv')
```

```python
pokemon.columns
```

```python
sns.barplot(x='isLegendary',y='Speed',data=pokemon)
plt.show()
```

```python
sns.barplot(x='isLegendary',y='Weight_kg',data=pokemon)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='isLegendary',y='Speed',hue='Generation',data=pokemon)
plt.show()
```

```python
sns.barplot(x='isLegendary',y='Weight_kg',data=pokemon,palette='Blues_d') #palette is pre-defined colour
plt.show()
```

```python
sns.barplot(x='isLegendary',y='Weight_kg',data=pokemon,palette='rocket') #palette is pre-defined colour
plt.show()
```

```python
sns.barplot(x='isLegendary',y='Weight_kg',data=pokemon,palette='vlag') #palette is pre-defined colour
plt.show()
```

```python
sns.barplot(x='isLegendary',y='Weight_kg',data=pokemon,color='orange') # give one color for all bars
plt.show()
```
